```python
#What is the average daily temperature range in this dataset?
```

```python
weather1.describe()
```

```python
temp_range = (weather1['Max TemperatureC'] - weather1['Min TemperatureC'])
average_temp_range = (temp_range).mean()
print("Average daily temperature range:", average_temp_range)
```

```python
#How would you visualize the trend of "Mean TemperatureC" over time?
```

```python
plt.figure(figsize=(10, 6))
plt.plot(weather1['CET'], weather1['Mean TemperatureC'])
plt.xlabel('CET')
plt.ylabel('Mean TemperatureC')
plt.title('Mean Temperature Trend Over Time')
```

#The plot shows a clear cyclic pattern, where temperatures rise and fall in a yearly cycle.

```python
#How would you plot a histogram of "Precipitationmm" to analyze rainfall distribution?
```

```python
val = weather1['Precipitationmm']
plt.figure(figsize=(10, 5))
plt.hist(val,bins=20)
plt.xlabel('Precipitation (mm)')
plt.ylabel('Frequency')
plt.title('Precipitationm')
plt.savefig('Histogram')
```

#This confirms that Madrid has a predominantly dry climate, with rainfall occurring in short, infrequent bursts.

```python
#How would you find the correlation between temperature and humidity levels?
```

```python
corr_btw_maxT_and_maxH = weather1['Max TemperatureC'].corr(weather1['Max Humidity'])
corr_btw_meanT_and_meanH = weather1['Mean TemperatureC'].corr(weather1[' Mean Humidity'])
corr_btw_minT_and_meanH = weather1['Min TemperatureC'].corr(weather1[' Min Humidity'])
print(corr_btw_maxT_and_maxH)
print(corr_btw_meanT_and_meanH)
print(corr_btw_minT_and_meanH)
```

#As temperature increases, humidity tends to decrease.

```python
correlations = {
    'Max Temp vs Max Humidity': weather1['Max TemperatureC'].corr(weather1['Max Humidity']),
    'Mean Temp vs Mean Humidity': weather1['Mean TemperatureC'].corr(weather1[' Mean Humidity']),
    'Min Temp vs Mean Humidity': weather1['Min TemperatureC'].corr(weather1[' Min Humidity'])
}
for key, value in correlations.items():
    print(f"{key}: {value}")

```
